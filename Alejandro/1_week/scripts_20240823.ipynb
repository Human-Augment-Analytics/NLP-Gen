## numpy

```python

import numpy as np
pdf_titles_by_group = np.array([
        ["title_1", "title_2"], 
        ["title_3","title_4"]
    ])
print(pdf_titles_by_group.shape)
print(pdf_titles_by_group.ndim)
print(pdf_titles_by_group.size)
print(pdf_titles_by_group.dtype)
```

## pandas

```python
import pandas as pd
df = pd.DataFrame(
    {
        "Group 1" : [
            "title_1", "title_2"
        ],
        "Group 2" : [
            "title_3","title_4"
        ]
    }
)
print(f" Group 2: {df['Group 2']}")
```

## matplotlib

```python
import matplotlib.pyplot as plt

pdf_titles = ["title_1", "title_2", "title_3", "title_4"]
values = [34, 23, 54,7]
plt.figure(1)
plt.suptitle('Research Papers')
plt.bar(pdf_titles, values)
plt.show()
```
